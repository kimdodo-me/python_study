```python
#칼럼 시프트
```

```python
samsung_1 = stock.get_market_ohlcv_by_date('20201018','20201118','005930')

#전날 거래량 만들기
samsung_1['전날거래량'] = samsung_1['거래량'].shift(1)

#전날 거래량 열 확인 
samsung_1
```

```python
#전날 거래량과 현재 거래량을 비교하는 조건문을 만들기

cond = samsung_1['거래량']>samsung_1['전날거래량']
samsung_1[cond]
```

```python
#전날 거래량에 비해 거래량이 증가한 일자 세기
print("거래일: " , len(samsung_1))
print('상승일: ',len(samsung_1[cond]) )
```

```python
#전날 종가에 비해 상승출발한 day세기 
cond = samsung_1['시가'] > samsung_1['종가']
samsung_1[cond]
#print('시초가 상승 day: ', len(samsung_1[cond]))
```

```python
#이동평균 데이터 구하기 
```

```python
#5일 이동평균선 구하기 
samsung_2 = stock.get_market_ohlcv_by_date('20200803','20201118','005930')

samsung_2['5일 이동평균'] = samsung_2['종가'].rolling( window = 5).mean()
#종가를 기준으로 pandas의 rolling을 이용 평균까지 구하기 
```

```python
print(samsung_2)
```

```python
#주어진 기간동안 시가가 5일 이동평균선을 돌파하는 횟수를 세어보기

#5일 이동평균이 당일의 종가를 합산하여 계산 된 것이기 때문에 다음날의 시가와 비교

cond = samsung_2['5일 이동평균'].shift(1) < samsung_2['시가']

#cond = samsung_2['5일 이동평균'] > samsung_2['시가'].shift(-1)

samsung_2[cond]
```

```python
print('거래일: ', len(samsung_2))
print('거래일: ', len(samsung_2[cond]))
```

```python
#데이터 샘플링 (data sampling)
```

```python
#월 단위로 데이터 샘플링

#frst() last()는 sampling시 대표값을 선정하는 기준이지 월초 월말이 아님 대신 데이터가 달라진다

mon = samsung_2.resample('M').last()
print(mon)
```

```python
#월별로 데이터를 수집하되 월초의 데이터로 수집하고 싶을 때

mons = samsung_2.resample('MS').first()
print(mons)
```

```python
#이렇게 월초의 데이터가 나오는데 월초의 행이 전부가 나오므로 월별 sampling이라 하기가 어렵다 


print(samsung_2)
```

```python
#실제 주가 데이터에서는 first last만으로 나타내기 어려움 시가는 시작가 종가는 종가 고가는 최고가등 주 월 단위의 특정 행 값이 필요한게 
#아님.

#apply 활용

how = {
    '시가':'first',
    '종가':'last',
    '고가':'max',
    '저가':'min',
    '거래량':'sum'
}

mon = samsung_2.resample('MS').apply(how)
print(mon)
```

```python
#수익률 계산 (pct_change())
```

```python
samsung_2['전날 대비 수익률'] = samsung_2['종가'].pct_change()*100
samsung_2
```

```python
#n일전 데이터로 부터 수익률을 구하고 싶을 때 

samsung_2['전날 대비 수익률'] = samsung_2['종가'].pct_change(periods=2)
samsung_2
```
