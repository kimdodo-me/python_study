```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

- **우선 자료가 필요하므로 20개의 정규분포 난수를 numpy ndarray로 생성한다.**

```python
np.random.seed(1000)
y = np.random.standard_normal(20)
y
```

- **plot의 인자로 그냥 ndarray값인 y를 넘겨도 자동으로 감지하고 처리하므로 x값을 지정해주지 않아도 된다**

```python
plt.plot(y)
```

```python
plt.plot(y.cumsum())
```

- **누적합을 나타내는 그래프**

- **matplotlib은 가독성을 높이기 위해 폰트를 바꿀 수도 있고 라벨을 붙이거나 grid를 추가해야 할 수 도있기 때문에 이를 위한 다양한 함수가 존재한다.**

```python
plt.plot(y.cumsum())
plt.grid(True) #그리드 추가
plt.axis('tight') # 축 간격을 더 세밀하게 조정
```

- **plt.axis의 옵션**
- off : 축선과 라벨을 없엔다
- equal : 가로축 세로축의 스케일을 같게 한다.
- scaled : 가로축과 세로축의 스케일이 같도록 크기 자체를 조정한다.
- tight : 축 범위 세밀하게 조정
- image : 모든 자료가 보이도록 축 범위를 자료의 극한 값으로 조정한다. 
- xlim,ylimx : 가로축, 세로축 범위 직접 조정
- ymin,ymax,xmin,xmax : 축 범위를 주어진 리스트 값으로 명시적으로 조정한다.

```python
plt.plot(y.cumsum())
plt.grid(True) 
plt.xlim(-1,20)
plt.ylim(np.min(y.cumsum())-1,np.max(y.cumsum())+1)
```

- **제목, x값 및 y 값을 설명하는 라벨을 추가 할 수 있다.**

```python
plt.figure(figsize = (7,4))
#figsize 인수를 사용하여 그림의 크기를 (폭,높이)로 정의한다.

plt.plot(y.cumsum(),'c',lw = 10) #lw는 선의 굵기 이고 'c'는 선의 색깔이네 그러면 두 저기 ro는 자연스럽게 점 색깔임
plt.plot(y.cumsum(),'ro')
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A simple plot')
```

- **표준 색깔 약자는 matplotlib 사이트가서 그때 그때 확인하자**

### 2차원 자료

 - **실생활에서는 대부분 2차원자료를 사용한다.**
 - **2차원 샘플 자료를 생성한다. 20*2크기의**

```python
np.random.seed(2000)
y = np.random.standard_normal((20,2)).cumsum(axis = 0)
y
```

```python
np.random.standard_normal((20,2)).cumsum(axis =0)
```

 - **일반적으로 plt.plot에는 2차원 배열도 인수로 넘길 수 있다. plt.plot명령은 자동으로 자료를 분석하여 담겨진 자료를 두개로 분리하고 그린다.**

```python
plt.figure(figsize = (7,4))
plt.plot(y, lw = 2)
#두 개의 선을 그린다.

plt.plot(y,'ro')
#두개의 점으로 된 선을 그린다

plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('a sample plot')
```

- **가독성을 높이기 위해 추가적인 주석기호를 달수도 있다. 각 자료에 라벨을 붙이고 범례에 이를 나열할 수 있다.**
- **plt.legend 명령은 범례의 위치를 지정하는 인수를 받는다. 인수값이 0이면 범례에 의해 가려져서 보이지 않는 자료의 수가 최소가 되도록 최적의 범례 위치를 자동으로 계산한다

```python
plt.figure(figsize = (7,4))
plt.plot(y[:,0], lw = 2,label = 'first')
plt.plot(y[:,1], lw = 2,label = 'second')
plt.plot(y,'ro')
plt.grid(True)
plt.legend(loc =0) #가능한 최적의 위치 
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('a sample plot')
```

- **동일한 금융 위험 요인을 여러 번 시뮬레이션하는 경우와 같이 비슷한 크기를 가진 여러개의 자료를 그릴 때는 하나의 y축을 그려도 되지만 크기가 전혀 다른 여러개의 자료를 그릴 때는 y축만 써서는 시각적인 정보를 제대로 표시하기 힘들다.** 

```python
y [:,1] = y[:,0]*100
plt.figure(figsize = (7,4))
plt.plot(y[:,0], lw = 2,label = 'first')
plt.plot(y[:,1], lw = 2,label = 'second')
plt.plot(y,'ro')
plt.grid(True)
plt.legend(loc =0) #가능한 최적의 위치 
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('a sample plot')
```

- **두번째 자료는 축으로 인해 거의 직선으로 보인다. 두번째 자료의 정보는 시각적으로 손실된 상태이다. 이를 해결하기 위한 방법은 두가지가 있다.** 

- **두 개의 y축을 사용하는 방법**
- **두 개의 서브플롯(subplot)을 사용하는 방법**

```python
fig, ax1 = plt.subplots() #첫 번째 자료를 첫 번째(왼쪽) 축에 대응하여 그린다.
plt.plot(y[:,0],'b', lw = 2,label = 'first')
plt.plot(y[:,0],'ro')
plt.grid(True)
plt.legend(loc =8) #중앙 아래 
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value first')
plt.title('a sample plot')
ax2 =  ax1.twinx() #두 번째 자료를 두 번째 축에 대응하여 그린다.
plt.plot(y[:,1],'g', lw = 2, label = 'second')
plt.plot(y[:,1],'ro')
plt.legend(loc = 0)
plt.ylabel('value second')
```

- **두 개의 subplot을 가지는 경우**

```python
plt.figure(figsize = (7,8))
plt.subplot(211)
plt.plot(y[:,0],lw = 2,label = 'first')
plt.plot(y[:,0],'ro')
plt.grid(True)
plt.legend(loc = 0)
plt.axis('tight')
plt.ylabel('value ')
plt.title('a sample plot1')
plt.subplot(212)
plt.plot(y[:,1],lw = 2,label = 'second')
plt.plot(y[:,1],'ro')
plt.grid(True)
plt.legend(loc = 0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('a sample plot2')
```

- **plt.subplot 함수는 numrows, numcols, fignum이라는 3개의 숫자를 인수로 받는다**
- **numrows는 행의수 numcols는 열의수 fignum은 몇 번째 서브플롯인지를 표시하는 수이다.**

- **예를들어 행과 열이 3개인 9개의 크기가 같은 서브플롯을 가진 그림에서 오른쪽 아래의 서브플롯은 plt.subplot(3,3,9)로 표시할 수 있다**

### scatterplot

- **스캐터 플롯은 자료 값중 하나를 x값으로 다른 값을 그에 대응하는 y 값으로 간주하는 것이다.**
- **예를 들어 두 개의 금융 시계열 자료의 수익률을 동시에 보일 때 스캐터 플롯을 사용할 수 있다.**

```python
y = np.random.standard_normal((1000,2))
plt.figure(figsize = (7,5))
plt.plot(y[:,0],y[:,1],'ro')
plt.grid(True)
plt.xlabel('first')
plt.ylabel('second')
plt.title('scatter plot')
```

- **matplotlib 라이브러리는 스캐터 플롯을 위한 별도의 scatter 함수도 제공한다. scatter함수는 plot함수와 기본적인 사용법이 같지만 몇가지 추가적인 기능을 제공한다.**

```python
plt.figure(figsize= (7,6))
plt.scatter(y[:,0],y[:,1],marker = 'o')
plt.grid(True)
plt.xlabel('first')
plt.ylabel('second')
plt.title('scatter plot ainnng')
```

- **scatter 함수는 세 번째 차원을 추가할 수 있으며 이 값은 여러 가지 색을 사용하여 시각화 할 수 있으며 색이 의미하는 칼라 바로 설명한다.**
- **0부터 10사이의 정수를 가지는 세 번째 자료를 임의로 생성**

```python
rand_num = np.random.randint(0,10,len(y))
len(rand_num)
```

```python
plt.figure(figsize= (7,6))
plt.scatter(y[:,0],y[:,1],c=rand_num,marker = 'o')
plt.colorbar()
plt.grid(True)
plt.xlabel('first')
plt.ylabel('second')
plt.title('scatter plot ainnng')
```

- **다른 유형은 히스토그램으로 주로 금융자산의 수익률을 나타내는데 사용된다.**

```python
plt.figure(figsize = (7,5))
plt.hist(y,label = ['first','second'],bins =20) #bins는 빈도 구분 값의 수를 의미한다. plt.hist의 인수도 matplotlib을 그때 그때 참고하자
plt.grid(True)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('histogram')
plt.legend()
```

- **두 자료에 대한 히스토그램**

```python
plt.figure(figsize = (7,4))
plt.hist(y,label = ['first','second'],color = ['b','g'],
        stacked = True, bins = 50)
plt.grid(True)
plt.legend()
plt.xlabel('value')
plt.ylabel('freq')
plt.title('hist')
```

- **박스 플롯 만들기 --> 주식 차트 만들때 유용하겠네**

```python
fig,ax = plt.subplots(figsize = (10,5))
plt.boxplot(y)
plt.setp(ax,xticklabels = ['first','second'])
plt.xlabel('data_set')
plt.ylabel('value')
plt.title('Boxplot')
```

## 금융 관련 Plot

- matplotlib 라이브러리에서는 특별한 금융관련 플롯도 일부 제공한다. 캔들스틱 차트와 같은 것들로 주로 과거의 주가 자료나 유사한 금융 시계열 자료를 시각화 하는데 쓰인다. 이러한 기능은 matplotlib.finance 하위 라이브러리에 있다. 

```python
import mplfinance as mpf
# 책에는 matplotlib.finance라고 되어있는데 mplfinance로 그냥 아예 대체됐나보다
```

- 함수를 사용할 때 해당하는 주식 기호와 시작일, 종료일만 지정하면 된다 
- ^GDAXI라는 독일 DAX지수 값을 가져온다. 라고 하고 싶었으나 mplfinance에서 quotes_historical_yahoo 없음

```python
from mpl_finance import candlestick2_ohlc
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
import pandas as pd
from pandas_datareader import data  
from datetime import datetime
from IPython.display import display
import FinanceDataReader as fdr
```

```python
start_date = datetime(2020,5,8)
end_date = datetime(2020,10,8)
```

```python
#fdr에서 apple 데이터 가져오기 
```

```python
aapl = fdr.DataReader('AAPL',start_date,end_date)
```

```python
aapl.head()
```

- **애플 이동평균선 데이터 그리기**

```python
aapl['MA3'] = aapl['Close'].rolling(3).mean()
aapl['MA5'] = aapl['Close'].rolling(5).mean()
aapl['MA10'] = aapl['Close'].rolling(10).mean()
aapl['MA20'] = aapl['Close'].rolling(20).mean()
display(aapl.head(5))
```

- **데이터 그리기**

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.set_title('AAPL', fontsize=15)
ax.set_ylabel("Price")
ax.set_xlabel("Date Time")
ax.plot(aapl.index, aapl[['Close','MA5','MA10']]) #여기서 x축은 index인 시간 y축은 종가(close)와 5일 10일 이동평균선
ax.legend(['Close','MA5','MA10'])
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
index = aapl.index.astype('str') # 캔들스틱 x축이 str로 들어감
#이동평균선 그리기
ax.plot(index, aapl['MA3'], label='MA3', linewidth=0.7) #캔들스틱의 x축은 날짜 데이터의 str(숫자형) 아마 epoch시간형식인듯
ax.plot(index, aapl['MA5'], label='MA5', linewidth=0.7)
ax.plot(index, aapl['MA10'], label='MA10', linewidth=0.7)

ax.xaxis.set_major_locator(ticker.MaxNLocator(20)) #티커 20개 제한 

# 그래프 title과 축 이름 지정
ax.set_title('AAPL', fontsize=22)
ax.set_xlabel('Date')

# 캔들차트 그리기
candlestick2_ohlc(ax, aapl['Open'], aapl['High'], 
                  aapl['Low'], aapl['Close'], # 시간 시가 고가 저가 종가 순
                  width=0.5, colorup='r', colordown='b')
ax.legend()
plt.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
top_axes = plt.subplot2grid((4,4), (0,0), rowspan=3, colspan=4)
bottom_axes = plt.subplot2grid((4,4), (3,0), rowspan=1, colspan=4, sharex=top_axes)
bottom_axes.get_yaxis().get_major_formatter().set_scientific(False) 


fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)


ax.bar(index, aapl['Volume'], color='r')



# 그래프 title과 축 이름 지정
ax.xaxis.set_major_locator(ticker.MaxNLocator(10))
ax.set_title('AAPL Volume', fontsize=22)
ax.set_xlabel('Date')

plt.grid()
plt.show()
```

```python
color_funct = lambda x: 'r' if x>=0 else 'b'
aapl['Volume'].diff().fillna(0) #na값들은 계산을 위해 0을 채워준다. 
color_df = aapl['Volume'].diff().fillna(0).apply(color_funct)
color_list = list(color_df) #구분된 색을 list형태로 만듦
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
ax.bar(index, aapl['Volume'], color=color_list)
ax.get_yaxis().get_major_formatter().set_scientific(False) # 거래량 값 그대로 표현
ax.set_title('AAPL Volume', fontsize=22)
ax.set_xlabel('Date')
plt.grid()
plt.show()
```

```python
# 그래프 구역 나누기
fig = plt.figure(figsize=(20,10))
top_axes = plt.subplot2grid((4,4), (0,0), rowspan=3, colspan=4)
bottom_axes = plt.subplot2grid((4,4), (3,0), rowspan=1, colspan=4, sharex=top_axes)
bottom_axes.get_yaxis().get_major_formatter().set_scientific(False) 


# 인덱스 설정
idx = aapl.index.astype('str')

#거래량 차트 그리기
color_fuc = lambda x : 'r' if x >= 0 else 'b'
color_list = list(aapl['Volume'].diff().fillna(0).apply(color_fuc))
bottom_axes.bar(idx, aapl['Volume'], width=0.5, 
                align='center',
                color=color_list)

#이동평균선 그리기
top_axes.plot(idx, aapl['MA3'], label='MA3', linewidth=0.7)
top_axes.plot(idx, aapl['MA5'], label='MA5', linewidth=0.7)
top_axes.plot(idx, aapl['MA10'], label='MA10', linewidth=0.7)

#캔들차트 그리기
candlestick2_ohlc(top_axes, aapl['Open'],aapl['High'], 
                  aapl['Low'], aapl['Close'],
                  width=0.5, colorup='r', colordown='b')

top_axes.set_title('AAPL', fontsize=22)
# X축 티커 숫자 20개로 제한
top_axes.xaxis.set_major_locator(ticker.MaxNLocator(20))
# X축 라벨 지정
bottom_axes.set_xlabel('Date', fontsize=15)

top_axes.legend()
plt.tight_layout()
plt.show()
```
